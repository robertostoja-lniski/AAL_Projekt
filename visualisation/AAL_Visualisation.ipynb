```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

df_raport = pd.read_csv('raport.csv', names=['operations'])
df_raport.head()
```

```python
df_times = pd.read_csv('time_raport.csv', names=['time'])
# to milisecs
df_times['time'] = df_times.time * 1000
df_times['time'] = df_times.time * df_times.time * df_times.time
df_times.head()
```

```python
frames = [df_raport, df_times]
df_full_raport = pd.concat(frames, axis = 1, sort = False)
```

```python
df_full_raport.sort_values("operations", ascending = True)
```

```python

X = df_full_raport.iloc[:, 0].values.reshape(-1, 1)
Y = df_full_raport.iloc[:, 1].values.reshape(-1, 1) 

linear_regressor = LinearRegression()
linear_regressor.fit(X, Y)
Y_pred = linear_regressor.predict(X)  

plt.scatter(X, Y)
plt.plot(X, Y_pred, color='red')
plt.rcParams["figure.figsize"] = (20,20)
plt.show()
```
